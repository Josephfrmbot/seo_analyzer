# Creating three lists  the good, the bad and the keywords

```python
good=[]
bad=[]
keywords=[]
```

# Analyzing the title

```python
title=soup.find('title').text
print(title)
```

# Analyzing the structure

```python
#The title

if title :
    good.append(f"Title Exists:{title}")
else:
    bad.append("Title Does Not Exist")
```

# The meta description

```python
#The Meta descritption

meta_d = soup.find('meta', attrs={'name': 'description'})['content']
meta_d
```

```python
if meta_d :
    good.append(f"Meta Exists:{meta_d}")
else:
    bad.append("Meta Does Not Exist")
```

```python
#The headings

hs = ['h1','h2','h3']
h_tags = []
```

```python
for h in soup.find_all(hs):
    good.append(f'{h.name}-->{h.text.strip()}')
    h_tags.append(h.name)
                
if 'h1' not in h_tags:
    bad.append("Heading Do Not Exist")
```

```python
# Image alt
for i in soup.find_all('img', alt=''):
    bad.append(f'No Alt:{i}')
```

```python
res = requests.get('https://www.notion.so/fr-fr/blog/baseline-project').text
# Parse the html of the url
soup = BeautifulSoup(res, 'html.parser')
```

```python
#Analyzing words within the body
bod = soup.find('body').text
```

```python
#Tokenizing the words
words = [i.lower() for i in word_tokenize(bod)]
```

```python
#Printing the first 30 words
words[:30]
```

# Erasing the stopwords

```python
sw= nltk.corpus.stopwords.words('french')
```

```python
# new_list without
new_words=[]
for i in words:
    if i not in sw and i.isalpha():
        new_words.append(i)
```

```python
# Show the list
new_words[:30]
```

```python
#Counting the most frequent words in the list
freq = nltk.FreqDist(new_words)
print(freq.most_common(10))
```
